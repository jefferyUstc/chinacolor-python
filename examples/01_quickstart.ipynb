```python
import sys
sys.path.insert(0, "..")
import seaborn as sns
import chinacolor as ctc
from matplotlib import pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'retina'
ctc.set_default_font_to_packaged()
```

# Quickstart
Basic usage of chinacolor: listing palettes, getting colors, and plotting.

```python
# List palettes (first 5)
pls = ctc.list_palettes()
print(f'In totoal, we have {len(pls)} palettes')
pls[-5:]
```

```python
# Get a palette as a list of hex colors
colors = ctc.get_palette('seq05', n=5)
colors
```

```python
# Plot the palette
ctc.plot_palette('seq08')
```

```python
# Use a colormap with matplotlib
cmap = ctc.get_cmap('div13', direction=-1)
Z = np.random.RandomState(0).randn(60, 60)
plt.imshow(Z, cmap=cmap)
plt.colorbar()
plt.title('Matplotlib with chinacolor colormap')
```
