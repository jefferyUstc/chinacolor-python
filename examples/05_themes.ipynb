```python
import sys
sys.path.insert(0, "..")
import seaborn as sns
import chinacolor as ctc
from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'retina'
ctc.set_default_font_to_packaged()
```

# Themes demo
Apply each chinacolor theme to seaborn/matplotlib.

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')

# Paper (auto categorical cycle)
with ctc.use_theme_ctc_paper(with_colors=True, color_cycle_n=8):
    ax = sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
    ax.set_title('CTC Paper Theme')
    plt.show()

# Mineral (auto categorical cycle)
with ctc.use_theme_ctc_mineral(with_colors=True):
    ax = sns.lineplot(data=tips, x='total_bill', y='tip')
    ax.set_title('CTC Mineral Theme')
    plt.show()

# Ink
with ctc.use_theme_ctc_ink():
    ax = sns.histplot(data=tips, x='total_bill', hue='day', multiple='stack')
    ax.set_title('CTC Ink Theme')
    plt.show()

# Bronze
with ctc.use_theme_ctc_bronze(oxidation_level='heavy'):
    ax = sns.kdeplot(data=iris, x='sepal_length', y='sepal_width', fill=True)
    ax.set_title('CTC Bronze Theme (heavy)')
    plt.show()

# Dunhuang
with ctc.use_theme_ctc_dunhuang(border_style='ornate'):
    ax = sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='petal_length', palette=ctc.get_cmap('seq05'))
    ax.set_title('CTC Dunhuang Theme')
    plt.show()
```
