```python
import sys
sys.path.insert(0, "..")
import seaborn as sns
import chinacolor as ctc
from matplotlib import pyplot as plt

%config InlineBackend.figure_format = 'retina'
ctc.set_default_font_to_packaged()
```

# Custom palettes by group/subgroup
Using `create_color_pick` and `custom_palette_pick` to build project-specific palettes.

```python
pick = ctc.create_color_pick(groups=[35, 27], subgroups=[[2, 4], [1, 3]], order_rule=1)
custom_colors = ctc.custom_palette_pick(pick, n=10)
custom_colors
```

```python
sns.set_theme(style='whitegrid')
sns.set_palette(custom_colors)
iris = sns.load_dataset('iris')
ax = sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
ax.set_title('Custom palette via group/subgroup selection')
plt.show()
```

# Custom palettes by ids / names / hex
