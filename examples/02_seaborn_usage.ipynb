```python
import sys
sys.path.insert(0, "..")
import seaborn as sns
import chinacolor as ctc
from matplotlib import pyplot as plt

%config InlineBackend.figure_format = 'retina'
ctc.set_default_font_to_packaged()
```

```python
ctc.set_theme('paper', with_colors=True, color_cycle_n=8)
```

# Seaborn categorical hue
Use a qualitative palette for categorical variables.

```python
tips = sns.load_dataset('tips')
ax = sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', palette=ctc.get_palette('qual15', n=6))
plt.show()
```

# Seaborn continuous hue
Use a sequential colormap for numeric variables.

```python
iris = sns.load_dataset('iris')
cmap = ctc.get_cmap('seq05', direction=1)
ax = sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='petal_length', palette=cmap)
ax.set_title('Continuous hue with chinacolor sequential colormap')
```

```python
# One-liner: pick recommended colormap for theme
cmap2 = ctc.pick_cmap_for_theme('paper', kind='sequential')
ax = sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='petal_length', palette=cmap2)
ax.set_title('Paper theme recommended sequential colormap')
plt.show()
```
