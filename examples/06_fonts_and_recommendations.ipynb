```python
import sys
sys.path.insert(0, "..")
import seaborn as sns
import chinacolor as ctc
from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'retina'
ctc.set_default_font_to_packaged()
```

# Packaged font
Enable the bundled Chinese font and apply a theme.

```python
fam = ctc.enable_packaged_chinese_font()
ctc.set_theme('paper', base_family=fam)
ax = sns.scatterplot(data=sns.load_dataset('iris'), x='sepal_length', y='sepal_width', hue='species')
ax.set_title(' -  (CTC Paper)')
plt.show()
```

# Theme recommendations
Get recommended palettes for a theme (categorical/sequential/diverging).

```python
rec = ctc.recommended_palettes_for_theme('dunhuang')
rec
```

```python
# Pick and use a recommended categorical palette
pal = ctc.pick_palette_for_theme('dunhuang', kind='categorical', n=6)
sns.set_palette(pal)
ax = sns.scatterplot(data=sns.load_dataset('tips'), x='total_bill', y='tip', hue='day')
ax.set_title('Dunhuang theme + recommended categorical palette')
plt.show()
```

```python
# Pick and use a recommended sequential colormap
cmap = ctc.pick_cmap_for_theme('dunhuang', kind='sequential', direction=1)
iris = sns.load_dataset('iris')
ax = sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='petal_length', palette=cmap)
ax.set_title('Dunhuang theme + recommended sequential colormap')
plt.show()
```
